```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder, MinMaxScaler, StandardScaler
```

```python
SP_500_stock = pd.read_csv('/content/drive/MyDrive/Datasets/Brazilian E-Commerce Public Dataset/S&P_500_stock data-all_stocks.csv')
olist_customers = pd.read_csv('/content/drive/MyDrive/Datasets/Brazilian E-Commerce Public Dataset/olist_customers_dataset.csv')
olist_geolocation = pd.read_csv('/content/drive/MyDrive/Datasets/Brazilian E-Commerce Public Dataset/olist_geolocation_dataset.csv')
olist_order_items = pd.read_csv('/content/drive/MyDrive/Datasets/Brazilian E-Commerce Public Dataset/olist_order_items_dataset.csv')
olist_order_payments = pd.read_csv('/content/drive/MyDrive/Datasets/Brazilian E-Commerce Public Dataset/olist_order_payments_dataset.csv')
olist_order_reviews = pd.read_csv('/content/drive/MyDrive/Datasets/Brazilian E-Commerce Public Dataset/olist_order_reviews_dataset.csv')
olist_orders_data = pd.read_csv('/content/drive/MyDrive/Datasets/Brazilian E-Commerce Public Dataset/olist_orders_dataset.csv')
olist_products = pd.read_csv('/content/drive/MyDrive/Datasets/Brazilian E-Commerce Public Dataset/olist_products_dataset.csv')
olist_sellers = pd.read_csv('/content/drive/MyDrive/Datasets/Brazilian E-Commerce Public Dataset/olist_sellers_dataset.csv')
product_category = pd.read_csv('/content/drive/MyDrive/Datasets/Brazilian E-Commerce Public Dataset/product_category_name_translation.csv')
```

```python
# Hợp nhất tất cả các dataframe cần thiết lại, bao gồm:
full_df = olist_orders_data.merge(olist_order_reviews, on='order_id')\
                           .merge(olist_order_payments, on='order_id')\
                           .merge(olist_customers, on='customer_id')\
                           .merge(olist_order_items, on='order_id')\
                           .merge(olist_products, on='product_id')\
                           .merge(olist_sellers, on='seller_id')

# Loại bỏ đi các bản ghi trùng lặp thông qua drop_duplicates()
full_df = full_df.drop_duplicates()
```

```python
#Xóa đi một số columns vì không có giá trị phân tích:
data = full_df.drop(columns = ['review_comment_title', 'review_comment_message', 'product_name_lenght',
                              'product_description_lenght', 'product_photos_qty'])

data = data.dropna()
data
```
